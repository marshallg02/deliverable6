```python
#step 1: importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# step 2: loading iris data
iris = sns.load_dataset("iris")
iris.head()
```

```python
#step 3: drop the "species" column
iris = iris.drop(labels = "species", axis = 1)
iris.head()
```

```python
#step 4: add 2 columns: sepal_sum and petal_sum
iris['sepal_sum'] = iris['sepal_length'] + iris['sepal_width']
iris['petal_sum'] = iris['petal_length'] + iris['petal_width']
iris.head()
```

```python
#step 5:
summary_stats = pd.DataFrame({
    'mean': iris.mean(),
    'std': iris.std(),
    'min': iris.min(),
    'max': iris.max()
})
```
